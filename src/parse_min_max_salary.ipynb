```python
import pandas as pd
```

```python
import json

# Assuming you have a JSON file named 'data.json' with some data
file_path = 'min_max_salary.json'

with open(file_path, 'r') as file:
    json_data = json.load(file)

# Now you can work with the 'json_data' variable, which contains the data from the JSON file
json_data
```

```python
df =pd.DataFrame.from_dict(json_data, orient='index')
```

```python
df.reset_index(inplace=True)
```

```python
df.rename(columns={'index': 'position'}, inplace=True)
```

```python
df.to_csv('salary_positions.csv', index=True)
```
